```python
import json 
from yelpapi import YelpAPI
import pandas as pd
```

```python
with open('D:\My Documents\GitHub\data-enrichment-wk14-activity-mapping-yelp-api-results\.secret\yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# Create an istance 
yelp_api = YelpAPI(login['api-key'], timeout_s = 5.0)
yelp_api
```

```python
results = yelp_api.search_query(location = 'NJ', term = 'Burger')
results.keys()
```

```python
burg = pd.DataFrame(results['businesses'])
burg.head()
```

```python
# total results 
results['total']
```

```python
results_2 = yelp_api.search_query(location = 'NJ', term = 'Burger', offset = 20)
results_2.keys()
```

```python
burg2 = pd.DataFrame(results_2['businesses'])
burg2.head()
```

```python
# concatenate 
burg_combined = pd.concat([burg, burg2], ignore_index = True)
display(burg_combined.head(3), burg_combined.tail(3))
```

```python
# save results 
burg_combined.to_json("D:/My Documents/Github/Stack-3/yelp-search-burger.json", orient = 'records')
```
